```python
def count_cars(video_path):
    cap = cv2.VideoCapture(video_path)
    car_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'cars.xml')

    line_position = 300
    count = 0

    while True:
        ret, frame = cap.read()
        if not ret:
            break
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        cars = car_cascade.detectMultiScale(gray, 1.1, 1)
        for (x, y, w, h) in cars:
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 0, 255), 2)
            if y > line_position and y < line_position + 5:
                count += 1
        cv2.imshow('Frame', frame)
        if cv2.waitKey(1) & 0xFF == ord('q'): # q tuşuna basarak döngüden çık
            break

    cap.release()
    cv2.destroyAllWindows()
    return count
```

```python
import cv2
import numpy as np

# Videoyu yükle
cap = cv2.VideoCapture("C:/Users/goksuk/Downloads/cars-traffic.mp4")

# Geçmiş kare
ret, frame1 = cap.read()
prvs = cv2.cvtColor(frame1, cv2.COLOR_BGR2GRAY)
hsv = np.zeros_like(frame1)
hsv[...,1] = 255

# Araç sayacı
car_count = 0

while True:
    # Yeni kareyi al
    ret, frame2 = cap.read()
    if not ret:
        break
    next = cv2.cvtColor(frame2, cv2.COLOR_BGR2GRAY)

    # Optik akış hesaplama
    flow = cv2.calcOpticalFlowFarneback(prvs, next, None, 0.5, 3, 15, 3, 5, 1.2, 0)

    # Hareketli bölgeleri tespit etme
    mag, _ = cv2.cartToPolar(flow[...,0], flow[...,1])
    hsv[...,0] = 0.8
    hsv[...,2] = cv2.normalize(mag, None, 0, 255, cv2.NORM_MINMAX)

    # Hareketli bölgeleri gösterme
    bgr = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)
    cv2.imshow('Frame', bgr)

    # Araçları sayma
    mask = cv2.inRange(hsv, np.array([0, 100, 100]), np.array([10, 255, 255]))
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    for cnt in contours:
        area = cv2.contourArea(cnt)
        if area > 200:
            cv2.drawContours(frame2, [cnt], -1, (0, 255, 0), 2)
            car_count += 1

    # Geçmiş kareyi güncelle
    prvs = next

    # Videoyu gösterme ve çıkış kontrolü
    cv2.imshow('Video', frame2)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Sonuçları göster
print("Toplam Araç Sayısı:", car_count)

# Temizleme
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# YOLOv4 modelini yükle
net = cv2.dnn.readNet("C:/Users/goksuk/Downloads/yolov4.cfg", "C:/Users/goksuk/Downloads/yolov4.weights")
layer_names = net.getLayerNames()
output_layers = net.getUnconnectedOutLayersNames()

# COCO veri setinden sınıf isimlerini yükleyin
with open("C:/Users/goksuk/Downloads/coco.names", "r") as f:
    classes = [line.strip() for line in f.readlines()]

# Videoyu açın
cap = cv2.VideoCapture("C:/Users/goksuk/Downloads/cars-traffic.mp4")

# Araç sayacı
car_count = 0

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    height, width, channels = frame.shape

    # YOLO'ya giriş için blob oluşturun
    blob = cv2.dnn.blobFromImage(frame, 0.00392, (416, 416), (0, 0, 0), True, crop=False)
    net.setInput(blob)
    outs = net.forward(output_layers)

    class_ids = []
    confidences = []
    boxes = []

    for out in outs:
        for detection in out:
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]
            if confidence > 0.5:
                center_x = int(detection[0] * width)
                center_y = int(detection[1] * height)
                w = int(detection[2] * width)
                h = int(detection[3] * height)
                x = int(center_x - w / 2)
                y = int(center_y - h / 2)
                boxes.append([x, y, w, h])
                confidences.append(float(confidence))
                class_ids.append(class_id)

    indexes = cv2.dnn.NMSBoxes(boxes, confidences, 0.5, 0.4)

    for i in range(len(boxes)):
        if i in indexes:
            x, y, w, h = boxes[i]
            label = str(classes[class_ids[i]])
            if label == "car":  # Sadece "car" sınıfını say
                car_count += 1
                color = (0, 255, 0)
                cv2.rectangle(frame, (x, y), (x + w, y + h), color, 2)
                cv2.putText(frame, label, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

    # Araç sayısını ekrana yazdırın
    cv2.putText(frame, 'Car Count: ' + str(car_count), (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)

    # Ekranı gösterin
    cv2.imshow("Frame", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Temizleme
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# YOLOv4 modelini yükle
net = cv2.dnn.readNet("C:/Users/goksuk/Downloads/yolov4.cfg", "C:/Users/goksuk/Downloads/yolov4.weights")
layer_names = net.getLayerNames()
output_layers = net.getUnconnectedOutLayersNames()

# COCO veri setinden sınıf isimlerini yükleyin
with open("C:/Users/goksuk/Downloads/coco.names", "r") as f:
    classes = [line.strip() for line in f.readlines()]


# Videoyu açın
cap = cv2.VideoCapture("C:/Users/goksuk/Downloads/cars-traffic.mp4")

# Araç sayacı ve izleyici listesi
car_count = 0
trackers = []
tracked_ids = set()

frame_id = 0

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    frame_id += 1

    height, width, channels = frame.shape

    # Her 10 karede bir YOLO'yu çalıştırarak yeni araçları tespit edin
    if frame_id % 10 == 0:
        # YOLO'ya giriş için blob oluşturun
        blob = cv2.dnn.blobFromImage(frame, 0.00392, (416, 416), (0, 0, 0), True, crop=False)
        net.setInput(blob)
        outs = net.forward(output_layers)

        class_ids = []
        confidences = []
        boxes = []

        for out in outs:
            for detection in out:
                scores = detection[5:]
                class_id = np.argmax(scores)
                confidence = scores[class_id]
                if confidence > 0.5:
                    center_x = int(detection[0] * width)
                    center_y = int(detection[1] * height)
                    w = int(detection[2] * width)
                    h = int(detection[3] * height)
                    x = int(center_x - w / 2)
                    y = int(center_y - h / 2)
                    boxes.append([x, y, w, h])
                    confidences.append(float(confidence))
                    class_ids.append(class_id)

        indexes = cv2.dnn.NMSBoxes(boxes, confidences, 0.5, 0.4)

        # Yeni araçlar için izleyiciler oluşturun
        for i in range(len(boxes)):
            if i in indexes:
                x, y, w, h = boxes[i]
                label = str(classes[class_ids[i]])
                if label == "car":  # Sadece "car" sınıfını izleyin
                    # Araç zaten izleniyorsa atla
                    if (x, y, w, h) not in tracked_ids:
                        tracker = cv2.TrackerCSRT_create()
                        trackers.append(tracker)
                        tracker.init(frame, (x, y, w, h))
                        tracked_ids.add((x, y, w, h))
                        car_count += 1

    # Tüm izleyicileri güncelleyin
    new_trackers = []
    for tracker in trackers:
        success, box = tracker.update(frame)
        if success:
            new_trackers.append(tracker)
            p1 = (int(box[0]), int(box[1]))
            p2 = (int(box[0] + box[2]), int(box[1] + box[3]))
            cv2.rectangle(frame, p1, p2, (0, 255, 0), 2, 1)
    trackers = new_trackers

    # Araç sayısını ekrana yazdırın
    cv2.putText(frame, 'Car Count: ' + str(car_count), (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)

    # Ekranı gösterin
    cv2.imshow("Frame", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Temizleme
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import dlib

# YOLOv4 modelini yükle
net = cv2.dnn.readNet("C:/Users/goksuk/Downloads/yolov4.cfg", "C:/Users/goksuk/Downloads/yolov4.weights")
layer_names = net.getLayerNames()
output_layers = net.getUnconnectedOutLayersNames()

# COCO veri setinden sınıf isimlerini yükleyin
with open("C:/Users/goksuk/Downloads/coco.names", "r") as f:
    classes = [line.strip() for line in f.readlines()]

# Videoyu açın
cap = cv2.VideoCapture("C:/Users/goksuk/Downloads/cars-traffic.mp4")

if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

# Araç sayacı ve izleyici listesi
car_count = 0
trackers = []

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    height, width, channels = frame.shape

    # Her 10 karede bir YOLO'yu çalıştırarak yeni araçları tespit edin
    if cap.get(cv2.CAP_PROP_POS_FRAMES) % 10 == 0:
        # YOLO'ya giriş için blob oluşturun
        blob = cv2.dnn.blobFromImage(frame, 0.00392, (416, 416), (0, 0, 0), True, crop=False)
        net.setInput(blob)
        outs = net.forward(output_layers)

        class_ids = []
        confidences = []
        boxes = []

        for out in outs:
            for detection in out:
                scores = detection[5:]
                class_id = np.argmax(scores)
                confidence = scores[class_id]
                if confidence > 0.5:
                    center_x = int(detection[0] * width)
                    center_y = int(detection[1] * height)
                    w = int(detection[2] * width)
                    h = int(detection[3] * height)
                    x = int(center_x - w / 2)
                    y = int(center_y - h / 2)
                    boxes.append([x, y, w, h])
                    confidences.append(float(confidence))
                    class_ids.append(class_id)

        indexes = cv2.dnn.NMSBoxes(boxes, confidences, 0.5, 0.4)

        # Yeni araçlar için izleyiciler oluşturun
        for i in range(len(boxes)):
            if i in indexes:
                x, y, w, h = boxes[i]
                label = str(classes[class_ids[i]])
                if label == "car":  # Sadece "car" sınıfını izleyin
                    tracker = dlib.correlation_tracker()
                    rect = dlib.rectangle(x, y, x+w, y+h)
                    tracker.start_track(frame, rect)
                    trackers.append(tracker)
                    car_count += 1

    # Tüm izleyicileri güncelleyin
    for tracker in trackers:
        tracker.update(frame)
        pos = tracker.get_position()
        start_x = int(pos.left())
        start_y = int(pos.top())
        end_x = int(pos.right())
        end_y = int(pos.bottom())
        cv2.rectangle(frame, (start_x, start_y), (end_x, end_y), (0, 255, 0), 2)

    # Araç sayısını ekrana yazdırın
    cv2.putText(frame, 'Car Count: ' + str(car_count), (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)

    # Ekranı gösterin
    cv2.imshow("Frame", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Temizleme
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2

# Cascade sınıflandırıcılarının dosya yolları
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
if face_cascade.empty():
    print("XML dosyası yüklenemedi!")

# Video akışı başlatma
cap = cv2.VideoCapture(0)  # 0, bilgisayarınızın kamerasını kullanır, farklı bir değer vererek video dosyası da kullanabilirsiniz

while True:
    # Kameradan bir kare al
    ret, frame = cap.read()
    if not ret:
        break

    # Gri tonlamaya dönüştür
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Yüzleri tespit et
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    # Yüzleri çerçeve içine al
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)

    # Sonuçları göster
    cv2.imshow('Face Detection', frame)

    # Çıkış için 'q' tuşuna basın
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Temizlik
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# HOG ve SVM modelini yükleme
hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())

# Video dosyasını açma
cap = cv2.VideoCapture("C:/Users/goksuk/Downloads/cars-traffic.mp4")  # Video dosyasının adını ve yolunu belirtin

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Yayaları tespit etme
    boxes, weights = hog.detectMultiScale(frame, winStride=(8, 8), padding=(32, 32), scale=1.05)

    # Yaya sayısını çerçeve üzerine yazma
    cv2.putText(frame, f'Yaya Sayısı: {len(boxes)}', (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    # Yayaları dikdörtgen içine alma
    for (x, y, w, h) in boxes:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

    # Çıktıyı gösterme
    cv2.imshow('Pedestrian Detection', frame)

    # 'q' tuşuna basıldığında döngüden çıkma
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Temizlik
cap.release()
cv2.destroyAllWindows()
```
